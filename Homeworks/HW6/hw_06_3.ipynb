```python
# Î”Â± ğ‘¡_(ğ›¼/2)âˆ— ğ‘†_Î” , Ğ³Ğ´Ğµ Î”= (ğ‘‹_1 )Â Ì…âˆ’ (ğ‘‹_2 )Â Ì…
# ğ·=1/2(ğ·_1+ğ·_2), Ğ³Ğ´Ğµ   ğ·_1 Ğ¸ ğ·_2 â€“ Ğ´Ğ¸ÑĞ¿ĞµÑ€ÑĞ¸Ğ¸ Ğ¾Ğ±ĞµĞ¸Ñ… Ğ³Ñ€ÑƒĞ¿Ğ¿
# ğ‘†_Î”=âˆš(ğ·/ğ‘›_1 +ğ·/ğ‘›_2 ), Ğ³Ğ´Ğµ ğ‘›_1 Ğ¸  ğ‘›_2 â€“ Ğ¾Ğ±ÑŠĞµĞ¼Ñ‹ Ğ²Ñ‹Ğ±Ğ¾Ñ€Ğ¾Ğº.
# ğ‘‘ğ‘“ = 2âˆ—(ğ‘›âˆ’1), Ğ³Ğ´Ğµ n- Ğ¾Ğ±ÑŠĞµĞ¼ 

# Ğ Ğ¾ÑÑ‚ Ğ´Ğ¾Ñ‡ĞµÑ€ĞµĞ¹ 175, 167, 154, 174, 178, 148, 160, 167, 169, 170
# Ğ Ğ¾ÑÑ‚ Ğ¼Ğ°Ñ‚ĞµÑ€ĞµĞ¹ 178, 165, 165, 173, 168, 155, 160, 164, 178, 175
# Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒÑ ÑÑ‚Ğ¸ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ¿Ğ¾ÑÑ‚Ñ€Ğ¾Ğ¸Ñ‚ÑŒ 95% Ğ´Ğ¾Ğ²ĞµÑ€Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ñ‹Ğ¹ Ğ¸Ğ½Ñ‚ĞµÑ€Ğ²Ğ°Ğ» Ğ´Ğ»Ñ Ñ€Ğ°Ğ·Ğ½Ğ¾ÑÑ‚Ğ¸ ÑÑ€ĞµĞ´Ğ½ĞµĞ³Ğ¾ Ñ€Ğ¾ÑÑ‚Ğ° Ñ€Ğ¾Ğ´Ğ¸Ñ‚ĞµĞ»ĞµĞ¹ Ğ¸ Ğ´ĞµÑ‚ĞµĞ¹.

import numpy as np
import scipy.stats as stats
```

```python
a = np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])
b = np.array([178, 165, 165, 173, 168, 155, 160, 164, 178, 175])

print(len(a), len(b))
```

```python
x_1 = np.mean(a)
x_2 = np.mean(b)
print(x_1, x_2)
```

```python
delta = abs(x_1 - x_2)
print(delta)        # Ğ¼Ğ¾Ğ¶ĞµÑ‚ Ğ»Ğ¸ Ğ´ĞµĞ»ÑŒÑ‚Ğ° Ğ±Ñ‹Ñ‚ÑŒ Ğ¾Ñ‚Ñ€Ğ¸Ñ†Ğ°Ñ‚ĞµĞ»ÑŒĞ½Ğ¾Ğ¹?
```

```python
d1 = np.var(a, ddof=1)
d2 = np.var(b, ddof=1)
d = (d1 + d2) / 2
print(d)
```

```python
se = np.sqrt(d / len(a) + d / len(b))
print(se)
```

```python
t = stats.t.ppf(0.975, 18)
print(t)
```

```python
l = delta - t * se
u = delta + t * se
print(l, u)
```
